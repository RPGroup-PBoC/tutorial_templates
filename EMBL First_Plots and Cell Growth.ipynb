Now, lets investigate how the number of ribosomes in E.coli relates to the doubling time and the dry mass of the cell. The goal here is to make different plots to try and find a simple relationship. Think about what a ribosome does. Think about how much dry mass a cell needs to produce to make two cells. Here is some data from Cell Bio by the Numbers: 

82A4B868-A644-458C-9AC2-535290CF0E5C.jpeg

```
#Data from Bionumbers on number of ribosomes in E.coli in thousands:
ribosomes = np.array([72,45,26,14,6.8])
#Data on doubling times in minutes
doubling_times = np.array([24,30,40,60,100])
#Data on dry mass in picograms (10^-12 grams)
dry_mass=([0.87,0.64,0.43,0.26,0.15])

plt.scatter(ribosomes, doubling_times)
plt.plot(ribosomes, doubling_times)

plt.xlabel('Number of ribosomes(thousands)')
plt.ylabel('Doubling time')
plt.show()
```

```
plt.scatter(dry_mass, ribosomes)
plt.plot(dry_mass, ribosomes)

plt.ylabel('Number of ribosomes(thousands)')
plt.xlabel('Cell mass (pg)')
plt.show()
```
