
This template provides the starting lines of code for the kinetic proofreading Gillespie simulation. Here, we begin to define the transition rates of the ribosome from one state to another according to the figure shown below (obtained by Chen *et al.* *The Biophysicist* 2020).

Below are the defined values for the kinetic rates in the Equilibrium model, where the off-rates of the wrong tRNA (states 2 or 4 to state 0) is higher than the corresponding off-rates of the correct tRNA (states 1 or 3 to state 0). In addition, for both correct and wrong tRNA, GTP synthesis (state 3 to state 1 and state 4 to state 2) is unlikely to occur and a GDP-bound tRNA is unlikely to bind to the free ribosome (state 0 to 3 or 4).

```python
# define rates of system, 'c' for correct, 'w' for wrong

# from Chan, Zuckerman, Nelson for an Equilibrium Model
kadd_c = 0.01
kadd_w = kadd_c

# fixed parameters for ribosome cycle - correct amino acid
# parameters based on in-vitro experimental measurements
kon_c = 40
koff_c = 0.5
lon_c = 0.26  # binding of the tRNA-GDP to satisfy detailed balance
loff_c = 0.085
mhc = 0.01
msc = 0.26    # synthesis of GTP to satisfy detailed balance
    
# fixed parameters for ribosome cycle - wrong amino acid
# in Hopfield-Ninio model, on rates are the same for correct and wrong, but we
# will increase the on rate of wrong tRNAs to see what happens
kon_w = kon_c
lon_w = lon_c
# off rate for wrong amino acid is higher than for correct
phi_off = 5; 
koff_w = phi_off*koff_c 
loff_w = phi_off*loff_c
# GTP hydrolysis rates are the same
mhw = mhc
msw = msc
```

Below, we have begun to enter the rates into a 5x5 matrix, where each column in the matrix corresponds to the current state of the ribosome and each row corresponds to the next state of the ribosome. However, we have left out some of the rates from the matrix for you to fill in below.

```python
# we need a 5x5 matrix for the 5 states of the ribosome
# rates_mat[i,j] will be the rate to go from state j to i
rates_mat = np.zeros((5,5))

# [0,0] is 0
rates_mat[1,0] = kon_c
rates_mat[2,0] = kon_w
rates_mat[3,0] = lon_c
rates_mat[4,0] = lon_w

# [1,1] [2,1] and [4,1] are all 0
rates_mat[0,1] = koff_c
rates_mat[3,1] = mhc

# [1,2], [2,2], and [3,2] are all 0
rates_mat[0,2] = 
rates_mat[4,2] = mhc

# [2,3], [3,3], and [3,4] are all 0
rates_mat[0,3] = loff_c + kadd_c
rates_mat[1,3] = msc

# [1,4], [3,4], and [4,4] are all 0
rates_mat[0,4] = 
rates_mat[2,4] = 
```
